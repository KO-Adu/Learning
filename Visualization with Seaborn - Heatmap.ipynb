## Visualization with Seaborn - Heatmap

https://www.youtube.com/watch?v=Gj-prU8aEgg

```python
import seaborn as sns
import numpy as np
%matplotlib inline
example = sns.load_dataset('flights')
example = example.pivot('month', 'year', 'passengers')
display = sns.heatmap(example)
```

```python
display = sns.heatmap(example, annot = True, fmt= 'd')
```

```python
display = sns.heatmap(example, cmap = "BuPu")
```

```python
display = sns.heatmap(example, linewidth = .10)
```
